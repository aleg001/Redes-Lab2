```python
import matplotlib.pyplot as plt
import seaborn as sns

similitudes = []

with open('similitudes.txt', 'r') as file:
    for line in file:
        similitudes.append(float(line.strip()))

plt.figure(figsize=(10, 6))
sns.histplot(similitudes, bins=20, kde=True)
plt.title("Histograma de Similitudes")
plt.xlabel("Similitud")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(similitudes, shade=True)
plt.title("Gráfico de Densidad de Similitudes")
plt.xlabel("Similitud")
plt.ylabel("Densidad")
plt.show()
```

```python
import pandas as pd
import numpy as np

similitudes = []

with open('similitudes.txt', 'r') as file:
    for line in file:
        similitudes.append(float(line.strip()))

df = pd.DataFrame(similitudes, columns=['Similitud'])

# Calcular estadísticas descriptivas
descripcion = df.describe()

# Calcular la mediana
mediana = df.median()

# Calcular la moda
moda = df.mode()

# Calcular la desviación estándar
std_dev = df.std()

print("Estadísticas Descriptivas:")
print(descripcion)
print("\nMediana:")
print(mediana)
print("\nModa:")
print(moda)
print("\nDesviación Estándar:")
print(std_dev)
```
